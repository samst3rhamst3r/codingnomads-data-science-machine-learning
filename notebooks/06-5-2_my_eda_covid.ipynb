```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('covid_19_data.csv')
```

```python
df.head()
```

```python
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])
```

```python
dfg = df.groupby('Country/Region')['Deaths'].sum().sort_values(ascending=False).head(10)
dfg
```

```python
ax = dfg.plot(kind='bar', title='COVID Rates by Country', xlabel='Country', ylabel='Number of Occurrences')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```
